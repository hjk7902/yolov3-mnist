## Face Mask Detection 데이터셋의 XML 파일의 아노테이션 형식 변환
* 변환 후 형식 : maksssksksss0.png 79,105,109,142,1 185,100,226,144,0 325,90,360,141,1

```python
import xml.etree.ElementTree as ET

xml_file = "maksssksksss29.xml"  

def xml2text(xml_file) :
    tree = ET.parse(xml_file)
    root = tree.getroot()

    result_text = ""

    filename = root.find("filename").text # 파일 이름 추출

    result_text += filename

    # 각 객체에 대한 정보 추출
    for obj in root.findall("object"):
        name = obj.find("name").text
        bndbox = obj.find("bndbox")
        if name=='with_mask' :
            label = 0
        elif name=='without_mask' :
            label = 1
        elif name=='mask_weared_incorrect' :
            label = 2

        xmin = int(bndbox.find("xmin").text)
        ymin = int(bndbox.find("ymin").text)
        xmax = int(bndbox.find("xmax").text)
        ymax = int(bndbox.find("ymax").text)

        result_text += f" {xmin},{ymin},{xmax},{ymax},{label}"
        
    return result_text
    # print(result_text) # maksssksksss0.png 79,105,109,142,1 185,100,226,144,0 325,90,360,141,1
```

## 내려받은 파일 일괄 처리

```python
import os

# 실제 디렉토리 경로로 바꿔주세요
directory = r"C:\Users\kosa\Downloads\archive\annotations"  

# 디렉토리 내의 파일 목록 얻기
file_list = os.listdir(directory)

# 파일 목록 출력
for file in file_list:
    print('data/mask/train/', xml2text(os.path.join(directory, file)), sep='')
```
